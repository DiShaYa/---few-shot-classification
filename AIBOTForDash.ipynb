```python
train_data = [
    ("Привет. Что ты умеешь?",  "Приветствие" ),
    ("Что ты делаешь?", "Приветствие"),
    ("Что такое ЗОЖ?", "Ответы"),
    ("Почему здоровый образ жизни считается ключевым фактором влияния на здоровье человека?", "Ответы"),]
```

```python
#сбор и токенизация предложений (вопросов) и приведение их в общи вид
inputs = tokenizer([x[0] for x in train_data], return_tensors="pt", truncation=True, padding=True)
#сбор категорий и приведение их приемлимый для обучения ии вид
labels = [x[1] for x in train_data]
unique_labels = list(set(labels))
label_dict = {v: k for k, v in enumerate(unique_labels)}
torch_labels = torch.tensor([label_dict[x] for x in labels])
```

```python
#обучение модели на основе функции потерь CrossEntropyLoss с помощью оптимизатора Adam
from torch.optim import Adam
from torch.nn import CrossEntropyLoss
#Оптимизатор Adam используется для настройки параметров модели во время обучения и задается скорость обучения
optimizer = Adam(model.parameters(), lr=1e-5)
#cоздание экземпляра для оценки функции потерь между ожидаемым результатом и истинными метками
loss_function = CrossEntropyLoss()

#В каждой эпохе модель будет обучаться на разных данных и обновлять свои веса
for _ in range(7):  # обучаем в  эпохах
    outputs = model(**inputs, labels=torch_labels)
    loss = loss_function(outputs.logits, torch_labels)
    loss.backward() #Вычисляются градиенты (относительную функцию потерь)
    optimizer.step()
    optimizer.zero_grad()

    print(f'Loss: {loss.item()}')
```

```python
# Вводимые вопросы
questions = [
    "Привет, что ты умеешь делать?",
              "Зачем ты нужен?",
              "Что такое ЗОЖ?",
              "Зачем нужен Зож?",]
```

```python
# Преобразовываем вопросы в формат, пригодный для модели
inputs = tokenizer(questions, return_tensors='pt', truncation=True, padding=True)

# Подаем вопросы в модель и получаем результаты
outputs = model(**inputs)
# Преобразовываем результаты в вероятности
probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
classes = ["Приветствие", "Ответы"]
#Получаем наиболее вероятные классы для каждого вопроса
predictions = [(classes[probs[i].argmax().item()], probs[i].argmax()) for i in range(len(probs))]
#Выводим пары вопрос - категория
for question,  prediction in zip(questions, predictions):
  print(f'Question: "{question}" - Predicted category: {prediction}, ')

#print('Вывод с вероятностями:')
#for question, prob in zip(questions, probs):
    #prob_list = prob.tolist()
    #prob_dict = {classes[i]: prob_list[i] for i in range(len(classes))}
    #print(f'Question: "{question}" - Probabilities: {prob_dict}')
```
